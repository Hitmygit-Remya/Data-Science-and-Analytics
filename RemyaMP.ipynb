# PL Plot

```python
plt.figure(figsize = (12, 9))
x = Myiris.PL
  
plt.hist(x, bins = 24, color = "blue")
plt.title("PL")
plt.xlabel("PL")
plt.ylabel("Count")
  
plt.show()
```
